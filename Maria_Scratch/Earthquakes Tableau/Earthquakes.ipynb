```python
# import dependencies
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
import json
import requests as req
import random
import seaborn as sns
import math as math
import time
from citipy import citipy
warnings.filterwarnings('ignore')
from pygeocoder import Geocoder

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# import files and upload them as dataframe
earthquakes_1 = pd.read_csv("Input/earthquakes.csv", encoding = "ISO-8859-1")
```

```python
earthquakes_1.head()
```

```python
#Keep selected columns
earthquakes = earthquakes_1[['Date','Time','Latitude','Longitude','Type','Depth','Magnitude','ID']]
earthquakes.head()
```

```python
#results = Geocoder.reverse_geocode(earthqaukes['Latitude'][0], earthqaukes['Longitude'][0])
# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
earthquakes['City'] = ""
earthquakes['Country'] = ""

count = 0
for index, row in earthquakes.iterrows():
    near_city = citipy.nearest_city(row['Latitude'], row['Longitude']).city_name
    near_country = citipy.nearest_city(row['Latitude'], row['Longitude']).country_code
    earthquakes.set_value(index,"City",near_city)
    earthquakes.set_value(index,"Country",near_country)
earthquakes.head()
```

```python
len(earthquakes)
```

```python
earthquakes['City']=earthquakes['City'].apply(lambda x: x.capitalize()) 
earthquakes['Country']=earthquakes['Country'].apply(lambda x: x.capitalize()) 
```
