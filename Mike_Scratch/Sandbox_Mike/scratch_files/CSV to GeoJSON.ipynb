```python
import pandas
import json
```

```python
data_frame = pandas.read_csv(
    '../web/public/data/historical_colours_world.csv',
    parse_dates=[
        'last_update',
        'input_date',
        'date_photo',
    ],
    infer_datetime_format=True,
    na_values=['']
)

data_frame[data_frame.columns[:10]].head()
```

```python
data_frame[data_frame.columns[20:]].head()
```

```python
data_frame.dtypes
```

```python
json_result_string = data_frame.to_json(
    orient='records', 
    double_precision=12,
    date_format='iso'
)
json_result = json.loads(json_result_string)
```

```python
geojson = {
    'type': 'FeatureCollection',
    'features': []
}
for record in json_result:
    geojson['features'].append({
        'type': 'Feature',
        'geometry': {
            'type': 'Point',
            'coordinates': [record['lng'], record['lat']],
        },
        'properties': record,
    })
```

```python
with open('../web/public/data/historical_colours_world.json', 'w') as f:
    f.write(json.dumps(geojson, indent=2))
```
